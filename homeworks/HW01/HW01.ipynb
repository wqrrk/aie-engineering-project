#### Первым делом импорт

```python
import numpy as np
from numpy import random as rn
```

#### Одномерные массивы

```python
#начнем с небольшого количества одномерных массивов
tArray = [2,4,6] 
b = np.array(tArray) # создание numpy ndarray из уже существующего массива
c = np.zeros(23) #создание 1-мерного массива из 23 нулей 
d = np.ones(5) #массив из 5 единиц
print("b: ", b) 
print("c: ", c)
print("d: ", d) 
```

#### Двумерные массивы

```python
#теперь создадим двумерные массивы и посмотрим, что вообще они умеют
matrix = np.arange(1,10).reshape(3,3) #делаем массив с числами от 1 до 9 и потом меняем его форму на матрицу 3 на 3
print("first matrix: \n", matrix) 
```

```python
tensor = np.ones(27).reshape(3,3,3)  #тензор, или же 3д массив 3x3x3 с единицами
for i in range(tensor.shape[0]): #идем по каждой оси(путешествуем в другие измерения xD), начиная с 0
    for j in range(tensor.shape[1]): 
        for k in range(tensor.shape[2]): 
            tensor[i,j,k] *= rn.randint(1,10) #просто умножим на любое число от 1 до 10
print("tensor: \n", tensor)  #получившаяся матрица 
```

```python
uMatrix = np.eye(3) #единичная матрица 
print("\nunity matrix: \n", uMatrix) 
```

#### Индексирование и срезы

```python
print(tensor)
```

### Еще немного срезов

```python
arr = np.array([
    [ 0,  1,  2,  3],
    [10, 11, 12, 13],
    [20, 21, 22, 23],
    [30, 31, 32, 33],
    [40, 41, 42, 43]
])
print("arr: \n", arr)
```

```python
print(arr[(2,3)]) #возьмем элемент из 2 строки 3 столбца(с нуля)
```

```python
print(arr[[2,3]]) #возьмем массив элементов со 2 по 3 строки
```

```python
print(arr[2:4,2]) #из строк со 2 по 4 (невкл) символы с индексом 2
```

#### Операции над массивами и немного статистики

```python
x = np.arange(0,10) 
y = np.arange(10,20) 
r1 = x + y 
r2 = x * y 
r3 = y // 3 
r4 = np.hstack((x,y)) 
r5 = np.vstack((x,y)) 
print(x, "\n", y, "\n")
```

```python
print("addition: \n",r1) 
print("multiplication: \n", r2) 
print("division: ", r3)
print("stacking horizontal: ", r4) 
print("stacking vertical: ", r5) 
```

```python
np.hsplit(r5,2) #разделим на две матрицы горизонтально - по столбцам
```

```python
np.vsplit(r5,2) #тоже разделим на две матрицы, но по строкам
```

```python
print(np.mean(r5)) #среднее арифметическое
```

```python
print(r5.min()) #минимум
print(r5.max()) #максимум
```
